```python
import pandas as pd
from sklearn.datasets import load_iris


```

```python
iris_data = load_iris()
features = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
features.head(15)
```

```python
iris_data.target
```

```python

#feature engineering: normalization
from sklearn.preprocessing import normalize
features = normalize(features)
features
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, iris_data.target, random_state=42)
```

```python
from keras.utils import to_categorical
y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)
y_train_cat
```

### Model Development

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(units=64, input_dim=4, activation='relu'))         # Input Layer
model.add(Dense(units=64, activation='relu'))                      # First Hidden Layer
model.add(Dense(units=3, activation='softmax'))                   # Output Layer
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train_cat, epochs=100, batch_size=10, verbose = 1, validation_data= (X_test, y_test_cat))
```

```python
# plot the history of the training
import matplotlib.pyplot as plt

# summarize the history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'], loc='upper left')
plt.show()
```

```python
# summarize the history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','test'], loc='upper left')
plt.show()
```

```python
from keras.datasets import fashion_mnist

#Map the data into test, train images and labels
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
#print the shape of the train image set
print(train_images.shape)

#print the shape of the test image set
print(test_images.shape)

#print the shape of the train label set
print(train_labels.shape)

#print the shape of the test label set
print(test_labels.shape)

#print the length of the train image set
print(len(train_labels))

#print the length of the test image set
print(len(test_labels))
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

# Map each label to its corresponding class name
class_labels = [class_names[label] for label in train_labels]

# Create a Pandas DataFrame with two columns
labeled = pd.DataFrame({"count": train_labels, "class": class_labels})

# Group the DataFrame by class and count the number of occurrences
class_counts = labeled.groupby("class").count()
class_counts
```

```python
# Create a Plotly figure showing the first image in the train_images set
import plotly.express as px
fig = px.imshow(train_images[0], title="First image in the train_images set")

# Add a colorbar to the figure
fig.update_layout(coloraxis_colorbar=dict(title="Color Scale"), template= 'plotly_dark')

# Show the figure
fig.show()
```

```python

# one hot encode test and train target values
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
print(train_labels.shape)
print(test_labels.shape)
```

```python
#Normalizing the data
#Dividing by 255 which is the maximum number. This is to keep the pixel ranges between zero and one.
train_images = train_images / 255.0

test_images = test_images / 255.0
```

#### Model Training

```python
from keras.layers import Flatten
from keras.layers import BatchNormalization
from keras.layers import Dropout
```

```python
model = Sequential()
model.add(Flatten(input_shape=(28, 28)))          #Input shape of 28 by 28
model.add(BatchNormalization())                   #Add a normalization layer
model.add(Dense(150, activation='relu'))         #Add a dense layer
model.add(BatchNormalization())                  #Another normalization layer
model.add(Dropout(0.2))                        #Add a dropout layer
model.add(Dense(150, activation='relu'))        #Another dense layer
model.add(BatchNormalization())                  #Another normalization layer
model.add(Dropout(0.2))                           #Add a dropout layer
model.add(Dense(150, activation='relu'))        #Another dense layer
model.add(BatchNormalization())                  #Another normalization layer
model.add(Dense(10, activation='softmax'))
```

```python
#Complie the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])


#Fit the model
history = model.fit(
    train_images,    # input training data
    train_labels,    # output training data
    batch_size=50,   # mini-batch gradient descent size
    epochs=15,       # number of iterations over the entire training data
    verbose=2,        # what type of information is printed during training
    validation_data=(test_images, test_labels))
```

```python
model.summary()
```
